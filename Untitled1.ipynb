```python
import pandas as pd
df = pd.read_csv('insurance.csv')
df = df.drop(columns=['region'])
```

```python
print(df.columns)
```

```python
#renaming columns
new_column_names = {
    'age': 'Age',
    'sex': 'Sex',
    'bmi': 'BMI',
    'children': 'Children',
    'smoker': 'Smoker',
    'charge': 'InsuranceCharge'
}
```

```python
# Rename the columns
df = df.rename(columns=new_column_names)
```

```python
# Define a function to categorize BMI
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 24.9:
        return 'Normal Weight'
    elif 24.9 <= bmi < 29.9:
        return 'Overweight'
    else:
        return 'Obese'

# Create a new column 'BMI_category' based on 'bmi'
df['BMI_category'] = df['BMI'].apply(categorize_bmi)
```

```python
print(df.columns)
```

```python
has_duplicates = df.duplicated().any()

# Print the result
if has_duplicates:
    print("Duplicate rows are present in the dataset.")
else:
    print("No duplicate rows found in the dataset.")
```

```python
# Check for null (NaN) values
null_values = df.isnull().any()

# Print the results
if null_values.any():
    print("Null (NaN) values are present in the dataset.")
    print(null_values)
else:
    print("No null (NaN) values found in the dataset.")
```

```python
plt.boxplot(df['BMI'])
plt.title('Box Plot for BMI')
plt.xlabel('BMI')
plt.show()
```

```python
# Calculate the IQR
Q1 = df['BMI'].quantile(0.25)
Q3 = df['BMI'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers
filtered_df = df[(df['BMI'] >= lower_bound) & (df['BMI'] <= upper_bound)]
```

```python
# Create a box plot for the filtered data
plt.figure(figsize=(8, 6))
plt.boxplot(filtered_df['BMI'])
plt.title('Box Plot for BMI (After Outlier Removal)')
plt.xlabel('BMI')
plt.show()
```

```python
# Save the cleaned DataFrame to the CSV file
output_file = 'cleaned_insurance.csv'
filtered_df.to_csv(output_file, index=False)

print(f"Cleaned dataset saved to '{output_file}'.")
```
